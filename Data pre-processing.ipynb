```python
#import librosa
import os
import numpy as np
import scipy.io.wavfile
import scipy.signal

DATA_PATH = r"/home/rithomas/data/FULL"
OUTPUT_PATH = '/home/rithomas/data/test_preprocessed'
AUGMENT = False  # apply data augmentation -- takes very long!!

all_files = []
all_wavs = []

if AUGMENT:
    pitch_shifts = range(-2, 3)
    stretch_rates = [0.9, 0.95, 1., 1.05, 1.1]
else:
    pitch_shifts = [0]
    stretch_rates = 1.

sr = 16000

# adapt data folder as needed
for folder, subfolder, files in os.walk(DATA_PATH):
    print(folder)
    for file in files:
        if file.endswith(".wav"):
            print(file)
            all_files.append(file)
            _sr, wave = scipy.io.wavfile.read(os.path.join(folder, file))
            secs = int(len(wave)/_sr) # Number of seconds in signal X
            print(secs)
            samps = secs*sr     # Number of samples to downsample
            wave = scipy.signal.resample(wave, samps)
            print(len(wave))
            if AUGMENT:
                aug_set = []
                for shift in pitch_shifts:
                    for rate in stretch_rates:
                        aug_set.append(librosa.effects.time_stretch(librosa.effects.pitch_shift(wave, sr, shift), rate))
            else:
                aug_set = [wave]
            num_seconds = 4
            for elem in aug_set:
                # this windows into 4 second frames with 75% overlap
                last_start = (len(elem) // ((num_seconds-1)*sr) - 1) * (num_seconds-1)*sr
                print(last_start)
                for start in range(0, last_start, sr*(num_seconds-1)):
                    all_wavs.append(elem[start:(start+sr*num_seconds)])
                    
print(len(all_files), len(all_wavs))

all_wavs = np.asarray(all_wavs)
if AUGMENT:
    np.save(OUTPUT_PATH, all_wavs)
else:
    np.save(OUTPUT_PATH, all_wavs)
```
